Number of unique values for each feature.

```
def unique_by_column(df: pd.DataFrame):
  print("{:30.26}|  {}".format("Feature", "Unique_value_count"))
  print(('-'*50))
  for column in df.columns:
    print("{:30.26}|  {:>18}".format(column, len(pd.unique(df[column]))))
```

```
unique_by_column(df)
```
